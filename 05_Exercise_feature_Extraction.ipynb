## Task 1: 
Load the skimage built-in images 'coins' and represent the coins that have a eccentricity larger and lower than the average. 



```python
from skimage import data
import skimage as sk 

Icoins = data.coins()
sk.io.imshow(Icoins)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

from skimage import data, measure
from skimage.measure import label
from skimage.filters import threshold_otsu

# compute the region properties of the coins in the image
props = measure.regionprops(Icoins)

# compute the average eccentricity of the coins
eccentricities = [p.eccentricity for p in props]
avg_eccentricity = sum(eccentricities)/len(eccentricities)

# create a figure
fig, axs, = plt.subplots(1, 2, figsize=(10,5))

# Display the coins with eccentricity larger than the average
axs[0].imshow(Icoins, cmap='gray')
axs[0].set_title('Coins with eccentricity > avg')
for p in props:
    if p.eccentricity > avg_eccentricity:
        minr, minc, maxr, maxc = p.bbox
        rect = plt.Rectangle((minc, minr), maxc - minc, maxr - minr, fill=False, edgecolor='red', linewidth=2)
        axs[0].add_patch(rect)

# Display the coins with eccentricity lower than the average
axs[1].imshow(Icoins, cmap='gray')
axs[1].set_title('Coins with eccentricity < avg')
for p in props:
    if p.eccentricity < avg_eccentricity:
        minr, minc, maxr, maxc = p.bbox
        rect = plt.Rectangle((minc, minr), maxc - minc, maxr - minr, fill=False, edgecolor='blue', linewidth=2)
        axs[1].add_patch(rect)

plt.show()

 
```

## Task 3: 
Consider  the function 'binary_blobs' from skimage to generate binary random blobs. Explore the effect of parameters 'blob_size_fraction' and 'volume_fraction'. 

```python
from skimage import data

IG_01 = data.binary_blobs(length=256, blob_size_fraction=0.1,volume_fraction=0.3)
IG_02 = data.binary_blobs(length=256, blob_size_fraction=0.9,volume_fraction=0.3)
IG_03 = data.binary_blobs(length=256, blob_size_fraction=0.01,volume_fraction=0.3)

IG_22 = data.binary_blobs(length=256, blob_size_fraction=0.1,volume_fraction=0.9)
IG_33 = data.binary_blobs(length=256, blob_size_fraction=0.1,volume_fraction=0.1)

# Get some figures:
fig, axes = plt.subplots(1,5,figsize=(20,10))
ax = axes.ravel()
ax[0].imshow(IG_01)
ax[0].set_title(' Default size and volume ') 

ax[1].imshow(IG_02,cmap=plt.cm.jet)
ax[1].set_title('Big size - default volume') # set figure title

ax[2].imshow(IG_03,cmap=plt.cm.jet)
ax[2].set_title('Small size - default volume') 

ax[3].imshow(IG_22,cmap=plt.cm.jet)
ax[3].set_title('Default size- big volume') 

ax[4].imshow(IG_33,cmap=plt.cm.jet)
ax[4].set_title('Default size - small volume') 

```
